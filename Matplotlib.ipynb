```python
#Pie Charts with matplotlib

import matplotlib.pyplot as plt

labels=['Category A','Category B','Category C']
sizes=[20,35,45]

plt.figure(figsize=(8,8))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels=['A',' B','C']
data1=[3,2,5]
data2=[4,3,2]
data3=[2,5,3]
data4=[3,4,5]
data5=[2,3,4]

x=np.arange(len(labels))

plt.figure(figsize=(10,6))
plt.bar(x,data1,label='Data 1')
plt.bar(x,data2, bottom=data1, label='Data 2')
plt.bar(x,data3,bottom=np.array(data1)+np.array(data2),label='Data 3')
plt.bar(x,data4,bottom=np.array(data1)+np.array(data2)+np.array(data3),label='Data 4')
plt.bar(x,data5,bottom=np.array(data1)+np.array(data2)+np.array(data3)+np.array(data4),label='Data 5')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
#plt.xticks(x,labels)
plt.legend()
plt.show()
```

```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

set1=set(['A','B','C'])
set2=set(['B','C','D','E'])

plt.figure(figsize=(8,8))
venn2([set1,set2],('Set 1','Set 2'))
plt.title('Venn Diagram')
plt.show()
```

```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

set1=set(['A','B','C'])
set2=set(['B','C','D','E'])
set3=set(['D','E','H','I'])
plt.figure(figsize=(8,8))
venn3([set1,set2,set3],('Set 1','Set 2','Set3'))
plt.title('Venn Diagram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.random.normal(0,1,100)
y1=np.random.normal(1,1,100)
x2=np.random.normal(2,1,100)
y2=np.random.normal(2,1,100)

plt.figure(figsize=(10,6))
plt.scatter(x1,y1,c='blue',label='Group 1')
plt.scatter(x2,y2,c='red',label='Group 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dates=pd.date_range(start='2024-01-01', periods=10,freq='D')
values1=np.random.rand(10)
values2=np.random.rand(10)
values3=np.random.rand(10)

plt.figure(figsize=(10,6))
plt.plot(dates,values1,marker='o',label='Series 1')
plt.plot(dates,values2,marker='x',label='Series 2')
plt.plot(dates,values3,marker='x',label='Series 3')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Line Graph')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dates=pd.date_range(start='2024-01-01', periods=10,freq='D')
data1=np.random.rand(10)
data2=np.random.rand(10)
data3=np.random.rand(10)

plt.figure(figsize=(10,6))
plt.stackplot(dates,data1,data2,data3,labels=['Data 1','Data 2','Data 3'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Stacked Area Chart')
plt.legend(loc='upper left')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

tasks=pd.DataFrame({
    'Task':['Task1','Task2','Task3'],
    'Start':['2024-01-01','2024-01-05','2024-01-10'],
    'End':['2024-01-03','2024-01-08','2024-01-15',]
})

fig=px.timeline(tasks,x_start='Start',x_end='End',y='Task',title='Gantt Chart')
fig.update_yaxes(categoryorder='total ascending')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dates=pd.date_range(start='2024-01-01',periods=10,freq='D')
categories=['A','B','C','D']
data=np.random.rand(10,4)
df=pd.DataFrame(data,index=dates,columns=categories)

plt.figure(figsize=(10,6))
sns.heatmap(df,annot=True,cmap='YlGnBu')
plt.xlabel('Category')
plt.ylabel('Date')
plt.title('Heat Map')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dates=pd.date_range(start='2024-01-01',periods=10,freq='D')
values=np.sin(np.linspace(0,2*np.pi,10))

plt.figure(figsize=(10,6))
plt.plot(dates,values,marker='o')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot')
plt.grid(True)
plt.show()
```
