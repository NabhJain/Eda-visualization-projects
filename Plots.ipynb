```python
import geopandas as gpd
import matplotlib.pyplot as plt
world= gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world['scaled']= world['pop_est']/10**6
world.plot(figsize=(10,10), column='scaled', legend=True)
plt.title('Cartogram of the world by Population')
plt.show()
```

```python
import folium
from folium.plugins import HeatMap
m= folium.Map(locations=[45.5236,-122.6750],zoom_start=5)
data=[
    [45.5236, -122.6750],
    [46.5236, -123.6750],
    [47.5236, -124.6750],
    [48.5236, -125.6750]
]
HeatMap(data).add_to(m)
m.save('heatmap.html')
```

```python
import matplotlib.pyplot as plt
labels= ['A','B','C','D']
sizes=[20,30,40,10]
fig, ax= plt.subplots()
wedges,texts,autotexts= ax.pie(sizes, labels=labels,
                               autopct='%1.1f%%', startangle=140,
                               wedgeprops= dict(width=0.3))

centre_circle= plt.Circle((0,0),0.2,color='white', fc='white')
fig.gca().add_artist(centre_circle)
ax.axis('equal')
plt.title('Ring Chart')
plt.show()
```

```python
import plotly.graph_objects as go
labels=['Root', 'A','B','C','A1','A2','B1','C1']
parents=['','Root','Root','Root','A','A','B','C']
fig= go.Figure(go.Sunburst(
    labels=labels,
    parents=parents,
    values=[10,20,30,10,10,10,10,10],
))
fig.update_layout(margin=dict(t=0, l=0, r=0, b=0))
fig.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
G= nx.DiGraph()
edges=[('Root','A'),('Root','B'),('A','A1'),('A','A2'),('B','B1')]
G.add_edges_from(edges)
pos=nx.spring_layout(G,seed=12)
plt.figure(figsize=(8,6))
nx.draw(G,pos,with_labels=True, arrows=True, node_size=2000,
        node_color='skyblue', font_size=10,font_weight='bold')
plt.title('Free Diagram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
data1= np.random.normal(0,1,1000)
data2= np.random.normal(2,1.5,1000)
plt.figure(figsize=(10,6))
plt.hist(data1 ,bins=30, alpha=0.5, label='Data 1')
plt.hist(data2 ,bins=30, alpha=0.5, label='Data 2')
plt.title('Multi-Dimensional Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
